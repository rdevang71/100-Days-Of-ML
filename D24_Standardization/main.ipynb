```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df = df.iloc[:,2:]
df
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.3,random_state=0)   
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)
X_test = pd.DataFrame(X_test)


X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test) 
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns= X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns= X_test.columns)
```

# Effect of  scaling

```python
#Befor
plt.scatter(X_train['Age'],X_train['EstimatedSalary'])
```

```python
# After
plt.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'],color = 'red')
```
